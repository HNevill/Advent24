```python


def part1(ans, nu):
    to_test = list(product(*(['*+='] * (len(nu)-1))))
    for j in range(len(to_test)):
        value = int(nu[0])
        for i in range(len(nu)-1):
            value = rec(value , int(nu[i+1]), to_test[j][i])
        if value == int(ans):
            return value



def rec(A,B,op):
    if op =='*':
        return A*B
    if op =='=':
        return int(str(A)+str(B))
    else:
        return A + B

```
