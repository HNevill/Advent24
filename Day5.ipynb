```python
test1, test2 = open(r'C:\Users\HannahNevill\.conda\Advent\Day5_2024').read().split("\n\n")
```

```python
import re
inputs = []
all_nu = []
for line in test1.splitlines():
    maps = (re.findall(r'(\d+)', line))
    inputs.append((maps[0], maps[1]))
    #print(maps)
    all_nu.append(maps[0])
    all_nu.append(maps[1])

all_nu = set(all_nu)
print(all_nu)
```

```python
myDict = {num: [] for num in all_nu}

for i in all_nu:
    for maps in inputs:
        if i == maps[0]:
            myDict[i].append(maps[1])
```

```python
print(myDict)
```

```python
accepted = []
nae_acpt=[]

for line in test2.splitlines():
    noobes = re.findall(r'(\d+)', line)
    #print(noobes)
    safe = True
    for i in range(len(noobes)-1):
        to_be_comp = myDict[noobes[i]]
        nu_ahead = noobes[i+1:len(noobes)]
        print(to_be_comp, nu_ahead)
        if all(elem in to_be_comp for elem in nu_ahead):
            print('pass')
        else:
            safe = False
    if safe == True:
        accepted.append(int(noobes[int((len(noobes)-1)/2)]))
    if safe == False:
        nae_acpt.append(noobes)
```
