```python
G = {i+j*1j: c for i,r in enumerate(open(r'C:\Users\HannahNevill\.conda\Advent\Day6'))
               for j,c in enumerate(r.strip())}
#enumerate gives the line numer i, and the content r. 
# j stores the index of the character, and c is the character itself

start = min(p for p in G if G[p] == '^')
block2=[]
def walk(G):
    pos, dir, seen = start, -1, set()
    while pos in G and (pos,dir) not in seen:
        seen |= {(pos,dir)}
        if G.get(pos+dir) == "#":
            dir *= -1j
        else: pos += dir


    return {p for p,_ in seen}, (pos,dir) in seen

path = walk(G)[0]
print(len(path),
      sum(walk(G | {o: '#'})[1] for o in path))
```
