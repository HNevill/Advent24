```python
def part1(A, B):
    diff1 = A[0]-B[0]
    diff2 = A[1]-B[1]
    loc1 = ((A[0]+diff1), (A[1] +diff2))
    loc2 = ((A[0]+(-2*diff1)),(A[1] +(-2*diff2)))
    return(loc1, loc2)
```

```python
def part2(A, B):
    diff1 = A[0]-B[0]
    diff2 = A[1]-B[1]
    antinode = []
    for i in range(1,52):
        antinode.append(((A[0]+(-(i)*diff1)),(A[1] +(-(i)*diff2)))) #starts at 2
        antinode.append(((A[0]+i*diff1), (A[1] + i*diff2))) #starts at 1
    return(antinode)
```
